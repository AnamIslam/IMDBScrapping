```python
movieTitle = []
releaseYear = []
certificate = []
runtime = []
genre = []
imdbRating = []
metascore = []
description = []
director = []
numVotes = []
gross = []

for movie in movie_all:
    
    #movie title
    m_title = movie.h3.a.text
    movieTitle.append(m_title)
    
    #release year
    m_ry = movie.find('span', class_= "lister-item-year text-muted unbold").text
    m_ry = m_ry[-5:-1]
    m_ry = int(m_ry)
    releaseYear.append(m_ry)
    
    #certificate
    m_cert_avail = movie.find('span', class_ = "certificate")
    if m_cert_avail == None:
        m_cert = "Can not catch"
        print("----------------------------------------------------------------")
    else:
        m_cert = m_cert_avail.text
    certificate.append(m_cert)
    
    #runtime
    m_rt_avail = movie.find('span', attrs = {'class':'runtime'})
    if m_rt_avail == None:
        m_rt = None
    else:
        m_rt = m_rt_avail.text

    runtime.append(m_rt)
    
    #genre
    m_g_avail = movie.find('span', attrs = {'class':'genre'})
    if m_g_avail == None:
        m_g = None
    else:
        m_g = m_g_avail.text
    genre.append(m_g)
    
    #imdbrating
    m_ir_avail = movie.strong
    if m_ir_avail == None:
        m_ir = None
    else:
        m_ir = movie.strong.text
    imdbRating.append(m_ir)
    
    #metascore
    if movie.find('div', class_ = "inline-block ratings-metascore") is not None:
        m_meta = movie.find('div', class_="inline-block ratings-metascore").span.text
    else:
        m_meta = "No Metascore found"
    metascore.append(m_meta)
    
    #description
    m_des_avail = movie.find_all('p', class_ = "text-muted")
    if len(m_des_avail) <2:
        m_des = ""
    else:
        m_des_avail = m_des_avail[1]
        if m_des_avail.find('a') != None and m_des_avail.find('a').text == "See full summary":  
            m_des_check = m_des_avail.find('a')['href']
            m_des_url = "https://www.imdb.com/"+m_des_check
            m_des_response = get(m_des_url, headers = headers)
            m_des_Soup = BeautifulSoup(m_des_response.text, 'html.parser')
            m_des_all = m_des_Soup.find('li', attrs={'class': 'ipl-zebra-list__item'})
            m_des = m_des_all.p.text
        else:
            m_des = m_des_avail.text;
            m_des = m_des[5:]
    #--------------------------------------------------------------------------------
    description.append(m_des)
    
    #director
    m_director_block_avail = movie.find('p', attrs = {'class': ''})
    m_total_director_block = m_director_block_avail.find('a')
    if m_total_director_block == None:
        m_director = None
    else:
        m_total_director = m_total_director_block.text
        m_total_director = m_total_director[:]
        for d in m_total_director_block.next_siblings:
            if d.name == 'span':
                break
            else:
                next_director = d.string
                next_director = next_director
                m_total_director = m_total_director + next_director
        m_director = m_total_director
    director.append(m_director)
    
    #numvotes
    m_nv_avail = movie.find('span', attrs= {'name':'nv'})
    if m_nv_avail == None:
        m_nv =None
    else:
        m_nv = movie.find('span', attrs= {'name':'nv'})['data-value']
    numVotes.append(m_nv)
    
    #gross
    m_g_avail = movie.find_all('span', attrs= {'name':'nv'})
    if len(m_g_avail) == 1:
        m_g=None
    else:
        m_g = movie.find_all('span', attrs= {'name':'nv'})[1]['data-value']
    gross.append(m_g)
    
    print(m_title+" "+m_director)
```

```python
#declearing all data we are gonna get
movieTitle = []
releaseYear = []
certificate = []
runtime = []
genre = []
imdbRating = []
metascore = []
description = []
director = []
numVotes = []
gross = []

#time -_- again
start_time = time()
requests = 0;
total_requests = len(years_url) *len(pages)

#for every year
for year_url in years_url:
    number_checker = 0
    for page in pages:
        
        #get request
        response = get("https://www.imdb.com/search/title/?title_type=feature&release_date="+year_url+"-01-01,"+year_url+
                       "-12-31&sort=num_votes,desc&count=250&start="+page+"&ref_=adv_nxt", headers = headers)
        
        #pause loop
        sleep(randint(3,7))
        
        #monitorr request
        requests+=1
        elapsed_time = time()-start_time
        print('Request: {}; Frequency:{} request/s'.format(requests, requests/elapsed_time))
        clear_output(wait=True)
        
        #throw warning if status code !=200
        if response.status_code!=200:
            warn('Request: {}; Status code: {}'.format(requests, response.status_code))
        if requests > total_requests:
            warn('Number of requests are greater than expected')
            break;
            
        #parse page content
        html_Soup = BeautifulSoup(response.text, 'html.parser')
        movie_all = html_Soup.find_all('div', class_="lister-item mode-advanced")
        
        for movie in movie_all:
            ###
             
            #movie title
            m_title = movie.h3.a.text
            movieTitle.append(m_title)
    
            #release year
            m_ry = movie.find('span', class_= "lister-item-year text-muted unbold").text
            m_ry = m_ry[-5:-1]
            m_ry = int(m_ry)
            releaseYear.append(m_ry)
    
            #certificate
            m_cert_avail = movie.find('span',attrs = {'class':'certificate'})
            if m_cert_avail == None:
                m_cert = None
                #print("----------------------------------------------------------------")
            else:
                m_cert = m_cert_avail.text
            certificate.append(m_cert)
    
            #runtime
            m_rt_avail = movie.find('span', attrs = {'class':'runtime'})
            if m_rt_avail == None:
                m_rt = None
            else:
                m_rt = m_rt_avail.text
            runtime.append(m_rt)
            
            #genre
            m_g_avail = movie.find('span', attrs = {'class':'genre'})
            if m_g_avail == None:
                m_g = None
            else:
                m_g = m_g_avail.text
            genre.append(m_g)
            
            #imdbrating
            m_ir_avail = movie.strong
            if m_ir_avail == None:
                m_ir = None
            else:
                m_ir = movie.strong.text
            imdbRating.append(m_ir)
            
            #metascore
            if movie.find('div', class_ = "inline-block ratings-metascore") is not None:
                m_meta = movie.find('div', attrs ={'class': 'inline-block ratings-metascore'}).span.text
            else:
                m_meta = None
            metascore.append(m_meta)
            
            #description
            m_des_avail = movie.find_all('p', attrs ={'class': 'text-muted'})
            if len(m_des_avail) <2:
                m_des = ""
            else:
                m_des_avail = m_des_avail[1]
                if m_des_avail.find('a') != None and m_des_avail.find('a').text == "See full summary":  
                    m_des_check = m_des_avail.find('a')['href']
                    m_des_url = "https://www.imdb.com/"+m_des_check
                    m_des_response = get(m_des_url, headers = headers)
                    sleep(randint(1,3))
                    m_des_Soup = BeautifulSoup(m_des_response.text, 'html.parser')
                    m_des_all = m_des_Soup.find('li', attrs={'class': 'ipl-zebra-list__item'})
                    m_des = m_des_all.p.text
                else:
                    m_des = m_des_avail.text;
                    m_des = m_des[5:]
            description.append(m_des)
            
            #director
            m_director_block_avail = movie.find('p', attrs = {'class': ''})
            m_total_director_block = m_director_block_avail.find('a')
            if m_total_director_block == None:
                m_director = None
            else:
                m_total_director = m_total_director_block.text
                m_total_director = m_total_director[:]
                for d in m_total_director_block.next_siblings:
                    if d.name == 'span':
                        break
                    else:
                        next_director = d.string
                        next_director = next_director
                        m_total_director = m_total_director + next_director
                m_director = m_total_director
            director.append(m_director)
            
            #numvotes
            m_nv_avail = movie.find('span', attrs= {'name':'nv'})
            if m_nv_avail == None:
                m_nv =None
            else:
                m_nv = movie.find('span', attrs= {'name':'nv'})['data-value']
            numVotes.append(m_nv)
            
            #gross
            m_g_avail = movie.find_all('span', attrs= {'name':'nv'})
            if len(m_g_avail) == 1:
                m_g=None
            else:
                m_g = movie.find_all('span', attrs= {'name':'nv'})[1]['data-value']
            gross.append(m_g)
            
            #print('Request: {}; Frequency:{} request/s'.format(requests, requests/elapsed_time))
            number_checker+=1
            print('Movie : {}; Year:{}; Request: {}; page: {}; movie: {};'.format(m_title, m_ry, requests, page,
                                                                                          number_checker,))
            
```

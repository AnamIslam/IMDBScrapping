```python
#declearing all data we are gonna get
movieTitle = []
director = []
releaseYear = []
certificate = []
runtime = []
genre = []
imdbRating = []
metascore = []
description = []
numVotes = []
gross = []

m_len = len(imdb_all_movie)

for m in range (0,m_len):
    
    #name
    tm_title = imdb_all_movie.loc[m,'Movies']
    movieTitle.append(tm_title)
    
    #director
    tm_director = imdb_all_movie.loc[m,'Director']
    tt = tm_director.replace("\n", "")
    ttt = tt.replace(", ", ",")
    director.append(ttt)
    
    #releaseYear
    tm_year = imdb_all_movie.loc[m,'Year']
    releaseYear.append(tm_year)
    
    #certificate
    tm_Certificate = imdb_all_movie.loc[m,'Certificate']
    if type(tm_Certificate) != str and np.isnan(tm_Certificate):
        tm_Certificate = 'Not Certified'
    certificate.append(tm_Certificate)
    
    #runtime
    tm_Runtime = imdb_all_movie.loc[m,'Runtime']
    if type(tm_Runtime)==float:
        ttt = -1
    else:
        tt = tm_Runtime[:-4]
        ttt = int(tt)
    runtime.append(ttt)
    
    #genre
    tm_Genre = imdb_all_movie.loc[m,'Genre']
    if type(tm_Genre) == float:
        ttt = "Not Given"
    else:
        tt = tm_Genre[1:]
        ttt = tt[:-12]
        ttt = ttt.replace(" ","")
    genre.append(ttt)
    
    #imdbRating
    tm_IMDB = imdb_all_movie.loc[m,'IMDB']
    imdbRating.append(tm_IMDB)
    
    #metascore
    tm_Metascore = imdb_all_movie.loc[m,'Metascore']
    if np.isnan(tm_Metascore):
        tm_Metascore = -1.1
    metascore.append(tm_Metascore)
    
    #description
    tm_Description = imdb_all_movie.loc[m,'Description']
    tt = tm_Description.replace(" ...                See full summary »",".")
    if "Add a Plot" in tt:
        print("-----"+ str(m))
        break
    description.append(tt)
    
    #numVotes
    tm_Votes = imdb_all_movie.loc[m,'Votes']
    numVotes.append(tm_Votes)
    
    #gross
    tm_Gross = imdb_all_movie.loc[m,'Gross'] #1197
    if type(tm_Gross)!= str and np.isnan(tm_Gross):
        tm_Gross = '-1'
    tt = tm_Gross.replace(',','')
    ttt = np.int64(tt)
    gross.append(ttt)
    
    print(m)

print(movieTitle)
    
```

```python
import pandas as pd
all_movie = pd.DataFrame({'Movies':movieTitle,
                          'Director':director,
                       'Year': releaseYear,
                       'Certificate': certificate,
                       'IMDB': imdbRating,
                       'Metascore': metascore,
                       'Votes': numVotes,
                       'Gross':gross,
                       'Runtime': runtime,
                       'Genre': genre,
                       'Description':description})
print(all_movie.info())
all_movie
```

```python
all_movie.to_csv('all_movie_rating_with_des_1951_2020_cleaned.csv')
print('Done')
```

```python
movieTitle = []
releaseYear = []
certificate = []
runtime = []
genre = []
imdbRating = []
metascore = []
description = []
numVotes = []
gross = []

for movie in movie_all:
    
    #movie title
    m_title = movie.h3.a.text
    movieTitle.append(m_title)
    
    #release year
    m_ry = movie.find('span', class_= "lister-item-year text-muted unbold").text
    m_ry = m_ry[-5:-1]
    m_ry = int(m_ry)
    releaseYear.append(m_ry)
    
    #certificate
    m_cert_avail = movie.find('span', class_ = "certificate")
    if m_cert_avail == None:
        m_cert = "Can not catch"
        print("----------------------------------------------------------------")
    else:
        m_cert = m_cert_avail.text
    certificate.append(m_cert)
    
    #runtime
    m_rt_avail = movie.find('span', attrs = {'class':'runtime'})
    if m_rt_avail == None:
        m_rt = None
    else:
        m_rt = m_rt_avail.text

    runtime.append(m_rt)
    
    #genre
    m_g = movie.find('span', class_ = "genre").text
    genre.append(m_g)
    
    #imdbrating
    m_ir = movie.strong.text
    imdbRating.append(m_ir)
    
    #metascore
    if movie.find('div', class_ = "inline-block ratings-metascore") is not None:
        m_meta = movie.find('div', class_="inline-block ratings-metascore").span.text
    else:
        m_meta = "No Metascore found"
    metascore.append(m_meta)
    
    #description
    m_des_avail = movie.find_all('p', class_ = "text-muted")
    if len(m_des_avail) <2:
        m_des = ""
    else:
        m_des = m_des_avail[1].text
    description.append(m_des)
    
    #numvotes
    m_nv = movie.find('span', attrs= {'name':'nv'})['data-value']
    numVotes.append(m_nv)
    
    #gross
    m_g_avail = movie.find_all('span', attrs= {'name':'nv'})
    if len(m_g_avail) == 1:
        m_g=None
    else:
        m_g = movie.find_all('span', attrs= {'name':'nv'})[1]['data-value']
    gross.append(m_g)
    
    print(m_title)
```

```python
#declearing all data we are gonna get
movieTitle = []
releaseYear = []
certificate = []
runtime = []
genre = []
imdbRating = []
metascore = []
description = []
numVotes = []
gross = []

#time -_- again
start_time = time()
requests = 0;

#for every year
for year_url in years_url:
    for page in pages:
        
        #get request
        response = get("https://www.imdb.com/search/title/?title_type=feature&release_date="+year_url+"-01-01,"+year_url+
                       "-12-31&sort=num_votes,desc&start="+page+"&ref_=adv_nxt", headers = headers)
        
        #pause loop
        sleep(randint(5,8))
        
        #monitorr request
        requests+=1
        elapsed_time = time()-start_time
        print('Request: {}; Frequency:{} request/s'.format(requests, requests/elapsed_time))
        clear_output(wait=True)
        
        #throw warning if status code !=200
        if response.status_code!=200:
            warn('Request: {}; Status code: {}'.format(requests, response.status_code))
        if requests > 105:
            warn('Number of requests are greater than expected')
            break;
            
        #parse page content
        html_Soup = BeautifulSoup(response.text, 'html.parser')
        movie_all = html_Soup.find_all('div', class_="lister-item mode-advanced")
        
        for movie in movie_all:
            ###
             
            #movie title
            m_title = movie.h3.a.text
            movieTitle.append(m_title)
    
            #release year
            m_ry = movie.find('span', class_= "lister-item-year text-muted unbold").text
            m_ry = m_ry[-5:-1]
            m_ry = int(m_ry)
            releaseYear.append(m_ry)
    
            #certificate
            m_cert_avail = movie.find('span', class_ = "certificate")
            if m_cert_avail == None:
                m_cert = None
                print("----------------------------------------------------------------")
            else:
                m_cert = m_cert_avail.text
            certificate.append(m_cert)
    
            #runtime
            m_rt = movie.find('span', class_ = "runtime").text
            runtime.append(m_rt)
            
            #genre
            m_g = movie.find('span', class_ = "genre").text
            genre.append(m_g)
            
            #imdbrating
            m_ir = movie.strong.text
            imdbRating.append(m_ir)
            
            #metascore
            if movie.find('div', class_ = "inline-block ratings-metascore") is not None:
                m_meta = movie.find('div', class_="inline-block ratings-metascore").span.text
            else:
                m_meta = None
            metascore.append(m_meta)
            
            #description
            m_des_avail = movie.find_all('p', class_ = "text-muted")
            if len(m_des_avail) <2:
                m_des = None
            else:
                m_des = m_des_avail[1].text
            description.append(m_des)
            
            #numvotes
            m_nv = movie.find('span', attrs= {'name':'nv'})['data-value']
            numVotes.append(m_nv)
            
            #gross
            m_g_avail = movie.find_all('span', attrs= {'name':'nv'})
            if len(m_g_avail) == 1:
                m_g=None
            else:
                m_g = movie.find_all('span', attrs= {'name':'nv'})[1]['data-value']
            gross.append(m_g)
            
            #print('Request: {}; Frequency:{} request/s'.format(requests, requests/elapsed_time))
            print('Movie Title: {}; Release year:{}; Request: {}' .format(m_title, m_ry, requests))
            
            
        
            
        
        
```
